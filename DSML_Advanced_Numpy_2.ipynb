*  Broadcasting is the interesting topic where a surtain array is broadcasted multiple times .




Screenshot (125).png

* Repeating behaviour is called broadcasting .
* In 1st example both arrays are of (4,3) and (4,3) size so we get output as addition of both.

* In 2nd example one array is of (4,3) size and another array is of (1,3) size so here, (1,3) size array is broadcased to (4,3) size .
So addition operation is possiable.

* In 3rd example we have two arrays of sizes (1,4) and (1,3).
* In this example both arrays are getting broadcasted . 
 (1,4) converted to (4,3) and (1,3) converted to (4,3).

```python
#Let's see implementation 

np.arange(0,40,10)
```

```python
#Let's stack such row one upon each other 

''' For that we use np.tile() function.
    Here (3,1) is how many times you want to repeate dimenssions.
    Where '3' means we have to repeate 1st dimenssion 3-times and '1'  means we have to repeate 2nd dimenssion 1 -time .'''

a = np.tile(np.arange(0, 40, 10), (3, 1))
```
