* It is an interesting feature of numpy.
* As we know numpy is written in C and it is optimized when you do 'y = x' it dose not allocate new memory for 'y'.

* RAM = [...............x1 (32 bytes)..y1 | (32 bytes) [0, 1, -75, 2]..]
* x = np.array([....])
* X-Metdata = x_id = epfjdsbjv3hg, 32 bytes, 1, 2, -75, 4
* x = X
* x = x_id
* X-Metdata = y_id = epfjoi3hg, 32 bytes, 1, 2, 3, 4


* So here we can see 'y' has assigned old x_id.
* There foe x = x_id = y.
* So when we change value of 'x' it will get reflected into 'y'
